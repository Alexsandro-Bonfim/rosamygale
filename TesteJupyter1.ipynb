```python
import pandas as pd
import ipywidgets as widgets
from ipywidgets import interact
import matplotlib.pyplot as plt

# Carregar o arquivo CSV usando o Pandas
url = "https://raw.githubusercontent.com/Alexsandro-Bonfim/rosamygale/main/Sales.csv"
df = pd.read_csv(url)

# FunÃ§Ã£o para exibir dados com base nas categorias selecionadas
def show_data(employee_id=True, employee_code=True, employee_name=True, pay_type=True, item_category=True):
    selected_columns = []
    
    if employee_id:
        selected_columns.append('Employee ID')
    if employee_code:
        selected_columns.append('Employee Code')
    if employee_name:
        selected_columns.append('Employee Name')
    if pay_type:
        selected_columns.append('Pay Type')
    if item_category:
        selected_columns.append('Item Category')
    
    # Exibir os dados selecionados
    display(df[selected_columns])

# Criar widgets interativos para seleÃ§Ã£o de categorias
employee_id_checkbox = widgets.Checkbox(value=True, description='Employee ID')
employee_code_checkbox = widgets.Checkbox(value=True, description='Employee Code')
employee_name_checkbox = widgets.Checkbox(value=True, description='Employee Name')
pay_type_checkbox = widgets.Checkbox(value=True, description='Pay Type')
item_category_checkbox = widgets.Checkbox(value=True, description='Item Category')

# Criar a interface interativa usando a funÃ§Ã£o interact
interact(show_data,
         employee_id=employee_id_checkbox,
         employee_code=employee_code_checkbox,
         employee_name=employee_name_checkbox,
         pay_type=pay_type_checkbox,
         item_category=item_category_checkbox)

# Adicionalmente, vocÃª pode adicionar grÃ¡ficos ou outros elementos visuais ao seu dashboard usando o matplotlib ou outras bibliotecas.
# Por exemplo, plotar um grÃ¡fico de barras para a contagem de categorias:
plt.figure(figsize=(10, 6))
df['Item Category'].value_counts().plot(kind='bar')
plt.title('Contagem de Item Categories')
plt.xlabel('Item Category')
plt.ylabel('Contagem')
plt.show()
```
